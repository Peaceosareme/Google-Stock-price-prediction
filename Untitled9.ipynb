```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta
today = date.today()

d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=1825)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2

data = yf.download('GOOG', 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
data["Date"] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True, inplace=True)
print(data.head())
```

```python
dep=data[['Date','Close']]
dep=dep.rename(columns={'Date':'ds','Close':'y'})

from fbprophet import Prophet
ph=Prophet()
ph.fit(dep)

predict = ph.make_future_dataframe(periods=730)
forcast = ph.predict(predict)

forcast[["ds", "yhat", "yhat_lower", "yhat_upper"]].tail()

```

```python
ph.plot(forcast, xlabel="Date", ylabel="Price")

# performance metrics for prophet
from fbprophet.diagnostics import cross_validation, performance_metrics
from fbprophet.plot import plot_cross_validation_metric

# perform cross-validation -- args: model, initial, period, horizon

df_cv = cross_validation(ph, initial='730 days', period='180 days', horizon = '365 days')
performance_metrics(df_cv)

plot_cross_validation_metric(df_cv, metric='rmse')

#mdape- median absolute percent error
#Median Absolute Percentage Error (MDAPE) is an error metric used to measure the performance of regression machine learning models. It is the median of all absolute percentage errors calculated between the predictions and their corresponding actual values. The resulting value is returned as a percentage which makes it easy to understand for end users.
plot_cross_validation_metric(df_cv, metric='mdape')
```

<h2> Prediction using Linear Regression

```python
x=data[['Open','High','Low']]
y=data[['Close']]
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
import statsmodels.api as sm
X=sm.add_constant(x)
model=sm.OLS(y,X)
results=model.fit()
print(results.summary())
```

```python
data.corr()# seeing the ols results and correlation, volume has a negative coreelation to other columns and p value aslo, advised to drop volume column
```

```python

x.head()
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=123)
lr.fit(x_train,y_train)
yha=lr.predict(x_test)
```

```python
mean_squared=mse(yha,y_test)
print('mean squared error ',mean_squared)

print('mean absolute error ',mae(yha,y_test))#mean absolute error

print('root mean squared error ', mean_squared ** 0.5)

```
